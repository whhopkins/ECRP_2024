```python
from ROOT import TFile, THStack, TCanvas, TLegend, gROOT, gStyle
from matplotlib import pyplot as plt
import numpy as np
gROOT.SetStyle("ATLAS");
gStyle.SetFillColor(10)
c = TCanvas('c')
c.SetLogy(True)
```

```python
file = TFile('StepHistograms.root')
file.cd('volumes/EMEC/stepProcess')
file.ls()
```

```python
hist_ep = file.Get('volumes/EMEC/stepProcess/EMEC_e+_vol_stepProcess')
hist_em = file.Get('volumes/EMEC/stepProcess/EMEC_e-_vol_stepProcess')
hist_gamma = file.Get('volumes/EMEC/stepProcess/EMEC_gamma_vol_stepProcess')
hist_neutron = file.Get('volumes/EMEC/stepProcess/EMEC_neutron_vol_stepProcess')
hist_nucleus = file.Get('volumes/EMEC/stepProcess/EMEC_nucleus_vol_stepProcess')
hist_other = file.Get('volumes/EMEC/stepProcess/EMEC_other_vol_stepProcess')
hist_pion = file.Get('volumes/EMEC/stepProcess/EMEC_pion_vol_stepProcess')
hist_proton = file.Get('volumes/EMEC/stepProcess/EMEC_proton_vol_stepProcess')
hist_list = [hist_ep, hist_em, hist_gamma, hist_neutron, hist_nucleus, hist_other, hist_pion, hist_proton]
```

```python
def sorted_hists(hist_list):
    ''' Sort histograms based on integral. Used further to stacking loop.'''
    draw_order=[]
    for k in sorted(hist_list, key=lambda x: x.Integral()):
        draw_order.append(k)
        
    return draw_order
```

```python
# color and stack
stack_hist = THStack('stack_hist','EMEC;Physics Process ID;Steps')
for i,hist in enumerate(sorted_hists(hist_list)):
    hist.SetLineColor(i+1)
    hist.SetFillColor(i+1)
    hist.SetLineWidth(2)
    stack_hist.Add(hist)
```

```python
# draw attributes
# Legent
leg = TLegend(0.80,0.65,0.95,0.9)
leg.SetTextFont(42)
leg.SetBorderSize(0)
leg.SetFillStyle(0)
for hist in reversed(sorted_hists(hist_list)):
    title = str(hist.GetTitle()).split('_')[1]
    leg.AddEntry(hist, title, "f")

# save
stack_hist.Draw()
leg.Draw()
c.SaveAs('EMEC_processes.pdf')
c.Clear()
```

```python
# processes frequency per particle type
for idx,histOfInterest in enumerate(hist_list):
    fig, ax = plt.subplots()
    process_freq_dict = {}
    for i in range(1, histOfInterest.GetNbinsX()+1):
        # process id: i-1, process counts: hist_gamma.GetBinContent(i)
        process_freq_dict[str(i-1)] = histOfInterest.GetBinContent(i)
    
    # remove items with zero value
    process_freq_dict = {k: v for k, v in process_freq_dict.items() if v>0}
    
    # Pie chart info
    labels = list(process_freq_dict.keys())
    sizes = list(process_freq_dict.values())
    
    ax.pie(sizes, labels=labels, autopct='%1.1f%%')
    ax.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
    ax.set(title=str(histOfInterest.GetTitle()).split('_')[1])

```
